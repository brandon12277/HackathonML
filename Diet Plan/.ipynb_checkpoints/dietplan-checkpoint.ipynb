```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("dietplan.csv")
df.head(10)
```

```python
df.shape
```

```python
df['activity_level'].unique()
```

```python
X = df.iloc[:,1:8].values
```

```python
X
```

```python
Y = df['calories_to_maintain_weight']
```

```python
Y
```

```python
X.shape
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()

ct = ColumnTransformer(transformers=[('ohe',ohe,[3])],remainder='passthrough')
```

```python
X_encoded = ct.fit_transform(X)
```

```python
Y.shape
```

```python
X_encoded.shape
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X_encoded, Y, test_size=0.2, random_state=42)
```

```python

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()

X_train[:, [1,2,4,5,6]] = sc.fit_transform(X_train[:, [1,2,4,5,6]])
X_test[:, [1,2,4,5,6]] = sc.transform(X_test[:, [1,2,4,5,6]])
```

```python
from sklearn.linear_model import LinearRegression

linreg = LinearRegression()

linreg.fit(X_train, Y_train)
```

```python
from sklearn.linear_model import BayesianRidge

brreg = BayesianRidge()

brreg.fit(X_train, Y_train)
```

```python

from sklearn.tree import DecisionTreeRegressor

dtreg = DecisionTreeRegressor()

dtreg.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score


score1 = linreg.score(X_test, Y_test)
score2 = brreg.score(X_test, Y_test)
score3 = dtreg.score(X_test, Y_test)

print("Linear Regression-> ",score1)
print("Bayesian Regression> ",score2)
print("Decision Tree Regression-> ",score3)
```

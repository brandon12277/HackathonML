```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib
```

```python
df=pd.read_csv('exercise_dataset.csv')
```

```python
df.head()
```

```python
df = df.drop('ID', axis=1)
```

```python
df.head()
```

```python
df['Gender']=df['Gender'].apply(lambda x: 1 if x=='Male' else 0)
```

```python
df.head()
```

```python
df['Exercise'] = df['Exercise'].str.slice(9)
```

```python
df.head()
```

```python
dummies=pd.get_dummies(df['Weather Conditions'])
```

```python
dummies
```

```python
df = df.drop('Weather Conditions', axis=1)
```

```python
df.head()
```

```python
fdf=pd.concat([df,dummies],axis='columns')
```

```python
fdf.head()
```

```python
fdf=fdf.drop('Sunny', axis=1)
```

```python
fdf.head()
```

```python
df['Exercise'] = df['Exercise'].astype(int)
```

```python
target=df['Exercise']
```

```python
target.head()
```

```python
x=fdf.copy()
x = x.drop('Exercise', axis=1)
```

```python
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,target,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```

```python
testset=[[343,70,71,20,1,20,148,28,4,1,0]]
```

```python
model.predict(testset)
```

```python
import pickle
with open('Exercise_model.pickle','wb') as f:
    pickle.dump(model,f)
```

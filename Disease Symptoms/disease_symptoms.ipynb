```python
X = df.iloc[:,1:7]
Y = df.Disease
```

```python
X.columns
```

```python
X.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in X.columns:
    if X[col].isnull().any():
        
        mask = X[col].notnull()
        X.loc[mask, col] = le.fit_transform(X.loc[mask, col])
    else:
        
        X[col] = le.fit_transform(X[col])

Y = le.fit_transform(Y)

X
```

```python
X.isnull().sum()
```

```python
from sklearn.impute import KNNImputer

cols_2_imp = ['Symptom_4', 'Symptom_5', 'Symptom_6']

imp=KNNImputer(n_neighbors=2);

X[cols_2_imp] = imp.fit_transform(X[cols_2_imp])
```

```python
X.isnull().sum()
```

```python
X
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train , Y_test = train_test_split(X,Y,test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

dtclas = DecisionTreeClassifier()

dtclas.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = dtclas.predict(X_test)

acc = accuracy_score(y_pred, Y_test)

print("Accuracy: ",acc)
```

```python
from joblib import dump

#LabelEncoder, model and knnimputer

dump(dtclas, 'model-disease.joblib')
dump(imp, 'knn-imputer.joblib')
dump(le, 'label-encode.joblib')
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("train.csv",  encoding='ISO-8859-1')
df
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df.shape
```

**CLEANING TEXTS**

```python
texts = df['selected_text']
texts.isnull().sum()
```

```python
from nltk import word_tokenize
import re
from nltk.corpus import stopwords
for i in range(len(texts)):
    print(texts[i])
    words = word_tokenize(texts[i])
    regex = r'\b(?!(?:#|@[A-Za-z0-9_]+|https?://)\b)\w+\b'
    words = [word.lower() for word in words if re.match(regex, word)]
    words = [word for word in words if word not in set(stopwords.words('english'))]
    texts[i] = ' '.join(words)
```

```python
texts
```

```python
target = df['sentiment']
target.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

target = le.fit_transform(target)
```

```python
from sklearn.model_selection import train_test_split


X_train, X_test, Y_train, Y_test = train_test_split(texts, target, test_size=0.2, random_state=42)

print("Train X")
print(type(X_train))
print("Test X")
print(X_test)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()

X_traint = tfidf_vectorizer.fit_transform(X_train)
X_testt = tfidf_vectorizer.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

reg = LogisticRegression(max_iter=1000)
reg.fit(X_traint, Y_train)
```

```python
from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier()
classifier.fit(X_traint, Y_train)
```

```python
from sklearn.naive_bayes import GaussianNB

naive = GaussianNB()
naive.fit(X_traint.toarray(), Y_train)
```

```python
y_pred = reg.predict(X_testt)


accuracy = accuracy_score(Y_test, y_pred)
print("Logistic Regresson Accuracy:", accuracy)
```

```python
y_pred = classifier.predict(X_testt)


accuracy = accuracy_score(Y_test, y_pred)
print("Classifier Randon ForestAccuracy:", accuracy)
```

```python
y_pred = naive.predict(X_testt.toarray())


accuracy = accuracy_score(Y_test, y_pred)
print("Naive Bayes Accuracy:", accuracy)
```
